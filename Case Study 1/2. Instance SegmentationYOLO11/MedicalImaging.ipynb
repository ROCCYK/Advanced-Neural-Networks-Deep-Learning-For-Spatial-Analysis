```python
import os
import numpy as np
import cv2
import torch
from ultralytics import YOLO
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score, jaccard_score
import matplotlib.pyplot as plt

# Paths to the test images and labels
test_images_folder = "/home/mohsin/School/Fall2024/Spatial_Analysis/Assignment3/brain_tumor/test/images/"
test_labels_folder = "/home/mohsin/School/Fall2024/Spatial_Analysis/Assignment3/brain_tumor/test/labels/"  

# Load the trained model
model = YOLO('/home/mohsin/School/Fall2024/Spatial_Analysis/Assignment3/brain_tumor/runs/segment/train6/weights/best.pt')

# Initialize metrics
total_metrics = {
    "precision": 0,
    "recall": 0,
    "accuracy": 0,
    "f1_score": 0,
    "dice": 0,
    "iou": 0,
    "pixel_accuracy": 0,
}

num_images = 0

# Iterate through all images in the test folder
for image_filename in os.listdir(test_images_folder):
    if not image_filename.endswith(".jpg"):
        continue

    # Load test image
    image_path = os.path.join(test_images_folder, image_filename)
    image = cv2.imread(image_path)
    height, width = image.shape[:2]

    # Run inference
    results = model(image_path)

    # Load corresponding ground truth label (YOLO format)
    label_filename = os.path.splitext(image_filename)[0] + ".txt"
    ground_truth_path = os.path.join(test_labels_folder, label_filename)
    
    if not os.path.exists(ground_truth_path):
        continue

    # Create ground truth mask from YOLO label
    ground_truth_mask = np.zeros((height, width), dtype=np.uint8)

    with open(ground_truth_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            parts = line.strip().split()
            if len(parts) < 5:
                continue

            class_id = int(parts[0])
            polygon_points = list(map(float, parts[5:]))
            if len(polygon_points) % 2 != 0:
                continue

            points = []
            for i in range(0, len(polygon_points), 2):
                x = int(polygon_points[i] * width)
                y = int(polygon_points[i + 1] * height)
                points.append([x, y])
            points = np.array(points, dtype=np.int32)
            cv2.fillPoly(ground_truth_mask, [points], 1)

    # Create binary mask from predicted output
    predicted_mask_img = np.zeros_like(ground_truth_mask)
    if results[0].masks is not None:
        for mask in results[0].masks.xyn:
            points = np.stack((mask[:, 0] * width, mask[:, 1] * height), axis=-1).astype(np.int32)
            cv2.fillPoly(predicted_mask_img, [points], 1)

    # Flatten masks for metric calculation
    ground_truth_flat = ground_truth_mask.flatten()
    predicted_flat = predicted_mask_img.flatten()

    # Calculate metrics for the current image
    precision = precision_score(ground_truth_flat, predicted_flat, zero_division=1)
    recall = recall_score(ground_truth_flat, predicted_flat, zero_division=1)
    accuracy = accuracy_score(ground_truth_flat, predicted_flat)
    f1 = f1_score(ground_truth_flat, predicted_flat, zero_division=1)
    dice = 2 * np.sum(predicted_flat * ground_truth_flat) / (np.sum(predicted_flat) + np.sum(ground_truth_flat) + 1e-6)
    iou = jaccard_score(ground_truth_flat, predicted_flat, zero_division=1)
    pixel_accuracy = np.sum(predicted_flat == ground_truth_flat) / len(ground_truth_flat)

    # Accumulate metrics
    total_metrics["precision"] += precision
    total_metrics["recall"] += recall
    total_metrics["accuracy"] += accuracy
    total_metrics["f1_score"] += f1
    total_metrics["dice"] += dice
    total_metrics["iou"] += iou
    total_metrics["pixel_accuracy"] += pixel_accuracy

    num_images += 1

# Calculate average metrics
average_metrics = {key: value / num_images for key, value in total_metrics.items()}

# Display results
for metric, value in average_metrics.items():
    print(f"{metric.capitalize()}: {value:.4f}")

# Plotting the metrics
plt.figure(figsize=(12, 6))
plt.bar(average_metrics.keys(), average_metrics.values(), color='red')
plt.xticks(rotation=45, ha="right")
plt.ylabel("Metric Value")
plt.title("Average Segmentation Metrics for YOLO Model on Test Set")
plt.tight_layout()
plt.show()
```

```python
# Applying augmentation to the train set and retraining the model

import os
import cv2
import numpy as np
from torch.utils.data import Dataset
import albumentations as A
from albumentations.pytorch import ToTensorV2

class TumorSegmentationDataset(Dataset):
    def __init__(self, image_dir, label_dir, output_image_dir, output_label_dir, transform=None):
        self.image_dir = image_dir
        self.label_dir = label_dir
        self.output_image_dir = output_image_dir
        self.output_label_dir = output_label_dir
        self.transform = transform
        self.image_paths = [os.path.join(image_dir, img) for img in os.listdir(image_dir) if img.endswith('.jpg')]

        # Create output directories if they don't exist
        os.makedirs(self.output_image_dir, exist_ok=True)
        os.makedirs(self.output_label_dir, exist_ok=True)

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        image = cv2.imread(image_path)
        height, width = image.shape[:2]

        # Load label and convert to polygon points
        label_path = os.path.join(self.label_dir, os.path.splitext(os.path.basename(image_path))[0] + ".txt")
        polygons = []
        with open(label_path, 'r') as f:
            for line in f.readlines():
                parts = list(map(float, line.strip().split()))
                if len(parts) < 5:
                    continue
                class_id = int(parts[0])
                polygon_points = [(parts[i] * width, parts[i + 1] * height) for i in range(1, len(parts), 2)]
                polygons.append((class_id, polygon_points))

        # Apply transformations to image and polygons
        if self.transform:
            transformed = self.transform(image=image, keypoints=[point for _, points in polygons for point in points])
            image = transformed['image']
            transformed_points = transformed['keypoints']

            # Restructure transformed points back into polygons
            start = 0
            transformed_polygons = []
            for class_id, points in polygons:
                num_points = len(points)
                transformed_polygons.append((class_id, transformed_points[start:start + num_points]))
                start += num_points

            # Convert transformed polygons back to YOLO format and save as label
            label_data = []
            for class_id, poly in transformed_polygons:
                poly_str = ' '.join(f"{x / width} {y / height}" for x, y in poly)
                label_data.append(f"{class_id} {poly_str}\n")

            # Save augmented label
            augmented_label_path = os.path.join(self.output_label_dir, f"aug_{os.path.basename(label_path)}")
            with open(augmented_label_path, 'w') as f:
                f.writelines(label_data)

            # Save augmented image
            augmented_image_path = os.path.join(self.output_image_dir, f"aug_{os.path.basename(image_path)}")
            cv2.imwrite(augmented_image_path, image.permute(1, 2, 0).numpy() * 255)  # Convert from tensor to image format

        return image
```
