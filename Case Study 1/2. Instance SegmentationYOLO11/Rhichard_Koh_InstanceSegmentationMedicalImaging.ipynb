```python
import os
from ultralytics import YOLO
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score, jaccard_score
from scipy.spatial.distance import directed_hausdorff

# Paths to the test images and labels
test_images_folder = "BRAIN-TUMOR.v1i.yolov11/test/images/"
test_labels_folder = "BRAIN-TUMOR.v1i.yolov11/test/labels/"

# Load the trained model
model = YOLO('runs/segment/train/weights/best.pt')

# Initialize metrics
total_precision = 0
total_recall = 0
total_accuracy = 0
total_f1 = 0
total_dice = 0
total_iou = 0
total_hausdorff = 0
total_mae = 0
total_pixel_accuracy = 0

num_images = 0

# Iterate through all images in the test folder
for image_filename in os.listdir(test_images_folder):
    if not image_filename.endswith(".jpg"):
        continue

    # Load test image
    image_path = os.path.join(test_images_folder, image_filename)
    image = cv2.imread(image_path)
    height, width = image.shape[:2]

    results = model(image_path)

    # Load corresponding ground truth label (YOLOv11 format)
    label_filename = os.path.splitext(image_filename)[0] + ".txt"
    ground_truth_path = os.path.join(test_labels_folder, label_filename)
    
    if not os.path.exists(ground_truth_path):
        continue

    # Create ground truth mask from YOLOv11 label
    ground_truth_mask = np.zeros((height, width), dtype=np.uint8)

    with open(ground_truth_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            parts = line.strip().split()

            # Ensure there are enough parts in each line (at least five for bounding box and additional points for polygon)
            if len(parts) < 5:
                print(f"Skipping line due to insufficient values: {line}")
                continue

            # Extract the class_id and the polygon points
            class_id = int(parts[0])

            # Extract polygon points (ignore the bounding box values for now)
            polygon_points = list(map(float, parts[5:]))

            # Ensure that we have pairs of points for (x, y) coordinates
            if len(polygon_points) % 2 != 0:
                print(f"Skipping line due to uneven number of polygon points: {line}")
                continue

            # Convert normalized coordinates to pixel coordinates
            points = []
            for i in range(0, len(polygon_points), 2):
                x = int(polygon_points[i] * width)
                y = int(polygon_points[i + 1] * height)
                points.append([x, y])

            # Convert points to numpy array
            points = np.array(points, dtype=np.int32)

            # Draw the polygon on the ground truth mask
            cv2.fillPoly(ground_truth_mask, [points], 1)

    # Create a binary mask from predicted output
    predicted_mask_img = np.zeros_like(ground_truth_mask)

    # Check if masks are present in the prediction results
    if results[0].masks is not None:
        for mask in results[0].masks.xyn:
            points = np.stack((mask[:, 0] * width, mask[:, 1] * height), axis=-1).astype(np.int32)
            cv2.fillPoly(predicted_mask_img, [points], 1)

    # Flatten masks for metric calculation
    ground_truth_flat = ground_truth_mask.flatten()
    predicted_flat = predicted_mask_img.flatten()

    # Calculate metrics for the current image
    precision = precision_score(ground_truth_flat, predicted_flat, zero_division=1)
    recall = recall_score(ground_truth_flat, predicted_flat, zero_division=1)
    accuracy = accuracy_score(ground_truth_flat, predicted_flat)
    f1 = f1_score(ground_truth_flat, predicted_flat, zero_division=1)
    dice = 2 * np.sum(predicted_flat * ground_truth_flat) / (np.sum(predicted_flat) + np.sum(ground_truth_flat) + 1e-6)  # Adding a small value to avoid division by zero
    iou = jaccard_score(ground_truth_flat, predicted_flat, zero_division=1)
    hausdorff_distance = directed_hausdorff(predicted_mask_img, ground_truth_mask)[0]
    mae = np.mean(np.abs(predicted_flat - ground_truth_flat))
    pixel_accuracy = np.sum(predicted_flat == ground_truth_flat) / len(ground_truth_flat)

    # Accumulate metrics
    total_precision += precision
    total_recall += recall
    total_accuracy += accuracy
    total_f1 += f1
    total_dice += dice
    total_iou += iou
    total_hausdorff += hausdorff_distance
    total_mae += mae
    total_pixel_accuracy += pixel_accuracy

    num_images += 1

# Calculate average metrics
if num_images > 0:
    average_precision = total_precision / num_images
    average_recall = total_recall / num_images
    average_accuracy = total_accuracy / num_images
    average_f1 = total_f1 / num_images
    average_dice = total_dice / num_images
    average_iou = total_iou / num_images
    average_hausdorff = total_hausdorff / num_images
    average_mae = total_mae / num_images
    average_pixel_accuracy = total_pixel_accuracy / num_images
else:
    print("No valid test images found.")
    exit()

# Plotting
metrics = {
    "Precision": average_precision,
    "Recall": average_recall,
    "Accuracy": average_accuracy,
    "F1 Score": average_f1,
    "Dice Coefficient": average_dice,
    "Jaccard Index (IoU)": average_iou,
    "Hausdorff Distance": average_hausdorff,
    "Mean Absolute Error (MAE)": average_mae,
    "Pixel Accuracy": average_pixel_accuracy
}

# Check metrics content
print("Metrics calculated: ", metrics)

# Plotting the metrics
plt.figure(figsize=(12, 6))
plt.bar(metrics.keys(), metrics.values(), color='blue')
plt.xticks(rotation=45, ha="right")
plt.ylabel("Metric Value")
plt.title("Average Segmentation Metrics for YOLOv11 Model on Test Set")
plt.tight_layout()

# Save the plot to an image file and display
plt.savefig("average_segmentation_metrics.png")
plt.show(block=True)
plt.close()
```
